```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
vol1_url = 'https://www.gutenberg.org/files/815/815-h/815-h.htm'
vol2_url = 'https://www.gutenberg.org/files/816/816-h/816-h.htm'
```

```python
soup = BeautifulSoup(requests.get(vol1_url).text, features='html')
chapter_dict = {}

chapter_tags = soup.find_all('div', class_='chapter')[1:]

for i, div in enumerate(chapter_tags):
    chapter_text = ''
    for p in div.find_all('p'):
        if 'class' not in p.attrs.keys():
            chapter_text += p.text
    if chapter_text == '':
        for p in chapter_tags[i+1].find_all('p'):
            if 'class' not in p.attrs.keys():
                chapter_text += p.text
    chapter_dict[div.h2.text] = chapter_text

dia1 = pd.DataFrame.from_dict(chapter_dict, orient='index').reset_index().rename(columns={'index':'chapter_title',0:'text'})
```

```python
import collections
soup = BeautifulSoup(requests.get(vol2_url).text, features='html')
chapter_dict = {}

for tag in soup.find_all('h2')[4:-58]:
    chapter_text = ''
    for t in tag.next_siblings:
        if t.name == 'p':
            chapter_text += t.text
    if 'Book' not in tag.text:
        chapter_dict[tag.text] = chapter_text

dia2 = pd.DataFrame.from_dict(chapter_dict, orient='index').reset_index().rename(columns={'index':'chapter_title',0:'text'})
```

```python
dia2
```

```python
dia = pd.concat([dia1,dia2]).reset_index(drop=True)
dia.to_csv('democracy_in_america_de_tocqueville.csv')
```
