```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
index_url = 'https://sabctrc.saha.org.za/documents/amntrans.htm'
```

```python
soup = BeautifulSoup(requests.get(index_url).text, features='html')
amnesty_dict = {}

for tr in soup.find('table', attrs={'class':'listtable'}).find_all('tr')[1:]:
    region = tr.find('td').text
    link = tr.a['href'].replace('..','https://sabctrc.saha.org.za')
    person_soup = BeautifulSoup(requests.get(link).text)
    for tr2 in person_soup.find('table', attrs={'class':'listtable'}).find_all('tr')[1:]:
        #amnesty_dict[tr2.find_all('td', {'class':'listdescrption'})[-1]] = (tr2.find('td'), region, tr2.find('td', {'class':'listdescrption'}), tr2.a['href'])
        if tr2.find_all('td')[3].text != '':
            amnesty_dict[tr2.find_all('td')[3].text.strip().title()] = (tr2.find('td').text.strip(), tr2.find_all('td')[2].text.strip(), region, tr2.a['href'].replace('../..','https://sabctrc.saha.org.za'))

trc = pd.DataFrame.from_dict(amnesty_dict, orient='index').reset_index().rename(columns={'index':'name',0:'date',1:'description',2:'region',3:'link'})
```

```python
def get_text(url):
    speaker_list = []
    soup = BeautifulSoup(requests.get(url).text, features='html')
    t = soup.find('div', {'class':'htmlcontainer'}).text
    return re.split(r'(?=\n[A-Z]+)',t)[1:]
```

```python
from tqdm import tqdm
tqdm.pandas()

trc['transcript_text'] = trc['link'].progress_apply(get_text)
```
